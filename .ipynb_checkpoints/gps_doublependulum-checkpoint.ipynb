```python
# import sys
# sys.modules[__name__].__dict__.clear()
from __future__ import division
import matplotlib.pyplot as plt
get_ipython().magic(u'matplotlib inline')
import numpy as np
import scipy as sp
import scipy.linalg
import time
import random
import tensorflow as tf
def print_np(x):
    print ("Type is %s" % (type(x)))
    print ("Shape is %s" % (x.shape,))
    print ("Values are: \n%s" % (x))
```

```python
import model
import cost
import iLQR
# model size
ix = 4;
iu = 2;
```

```python
import poliOpt
import trajOpt
```

```python
# Initial trajectory distribution from iLQR with low number of iterations
i1 = iLQR.iLQR('doublePendulum',200,5000)
x0 = np.zeros(4)
u0 = np.ones( (200,2) )
x0[0] = - np.pi/2
x0[1] = 0.
x,u, Quu_save, Quu_inv_save = i1.update(x0,u0)
```

```python
# policy, traj class
# del myPolicy
myPolicy = poliOpt.poliOpt('test',ix,iu,20,10000,0.1)
myTraj = trajOpt.trajOpt('test',200)
# Nu & Dual
nu = np.ones(200) * 0.01
dual = np.ones((200,2)) * 0.01
# parameters
maxIter = 1
```

```python
for i in range(maxIter) :
    # policy optimization start!!
    # data extraction
    x_data = x[0:200,:]
    u_data = u
    var_inv_data = Quu_save
    # dual initial value!!
    W1,b1,W2,b2,var = myPolicy.update(x_data,u_data,var_inv_data,dual)

    # trajectory optimization
    myTraj.setEnv(myPolicy,dual,nu)
    x_new,u_new, Quu_save_new, Quu_inv_save_new = myTraj.update(x0,u_data)

    # dual update
    alpha = 0.1
    u_traj = u_new
    u_poli = myPolicy.getPolicy(x_new[0:200,:])
    dual_new = dual + alpha * (u_poli - u_traj) * np.transpose(np.tile(nu,(2,1)))
```

```python
np.savetxt('w1.txt',W1)
np.savetxt('w2.txt',W2)
np.savetxt('b1.txt',b1)
np.savetxt('b2.txt',b2)
# plt.plot(Quu_save_new[:,1,1], linewidth=2.0)
```
